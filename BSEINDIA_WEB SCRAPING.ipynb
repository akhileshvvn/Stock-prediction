```python
from bs4 import BeautifulSoup
from urllib.request import Request,urlopen
```

```python
url = Request('https://www.bseindia.com/markets/equity/EQReports/StockPrcHistori.aspx?expandable=7&scripcode=500209&flag=sp&Submit=G',headers = {'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(url).read()
webpage
```

```python
soup = BeautifulSoup(webpage,'lxml')
type(soup)
```

```python
k = soup.find_all('div')
k
```

```python
lis=[]
temp=[]
n=0
count=0
for i in k:
    l = i.find_all('td')
    for j in l:
        if(j.getText()=='Date'):
            n=n+1
        if(n>0):
            if(j.getText()!='* Spread'):
                if(count<13):
                    temp.append(j.getText())
                    count=count+1
                else:
                    lis.append(temp)
                    temp=[]
                    temp.append(j.getText())
                    count=1
lis
```

```python
import pandas as pd
df = pd.DataFrame(lis)
df.drop(31) 
```

```python
df.columns = df.iloc[0,:]
df = df.drop(0)
df = df.drop(31)
df.head()
```
