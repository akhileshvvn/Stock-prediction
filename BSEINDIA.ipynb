```python
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request
```

```python
link = Request("https://www.bseindia.com/markets/equity/EQReports/StockPrcHistori.aspx?expandable=7&scripcode=500209&flag=sp&Submit=G",headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(link).read()
webpage
```

```python
soup = BeautifulSoup(webpage,'lxml')
type(soup)
```

```python
k=soup.find_all('tr')
k
```

```python
lis=[]
temp=[]
n=0
count=0
for i in k:
    l = i.find_all('td')
    for j in l:
        if(j.getText()=='Date'):
            n=n+1
        if(n>0):
            if(j.getText()!='* Spread'):
                if(count<13):
                    temp.append(j.getText())
                    count=count+1
                else:
                    lis.append(temp)
                    temp=[]
                    temp.append(j.getText())
                    count=1
lis
```

```python
import pandas as pd
df = pd.DataFrame(lis)
df.head() 
```

```python
df.columns = df.iloc[0,:]
df = df.drop(0)
df = df.drop(31)
df.head()
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
z = list(map(float,list(df['High'])))
date_objects = [datetime.strptime(date, '%d/%m/%y').date() for date in df['Date']]
plt.plot(date_objects,z)
plt.show()
```

```python
import numpy as np
dat = np.array(df['Open']).reshape(-1,1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dat,df['High'])
```

```python
y_pred = model.predict(np.array([700,600,800]).reshape(-1,1))
y_pred
```
